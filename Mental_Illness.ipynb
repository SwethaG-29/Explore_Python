### STEP 3: Understanding the Dataset

```python
### Writing a function for - name of columns,data type of columns, No of rows, unique values, Missing values 
def describe_datasets(df):

    title = []
    dtype = []
    count = []
    unique =[]
    missing_values = []

    for col in df.columns:
        title.append(col)
        dtype.append(df[col].dtype)
        count.append(len(df[col]))
        unique.append(len(df[col].unique()))
        missing_values.append(df[col].isna().sum())

    output = pd.DataFrame({
        "Title": title, "Dtype": dtype, "Count": count,"Unique values": unique, "missing_values": missing_values
    })
    return output
        
```

```python
# Adding colours for Heading/Descrption of File name 
class colours:
    BLUE = '\033[94m'
    BOLD = "\033[1m"
    UNDERLINE = "\033[4m"
    END = '\033[0m'
```

```python
### Giving borders to heading and calling function & printing it.
print(colours.BLUE+ colours.BOLD+colours.UNDERLINE+"The Description of df1:"+colours.END)
print(describe_datasets(df1))

print("\n",colours.BLUE+colours.BOLD+colours.UNDERLINE+"The Descrpition of df2:"+colours.END)
print(describe_datasets(df2))

print("\n", colours.BLUE+colours.BOLD+colours.UNDERLINE+"The Description of df3:"+colours.END)
print(describe_datasets(df3))

print("\n", colours.BLUE+colours.BOLD+colours.UNDERLINE+"The description of df4:"+colours.END)
print(describe_datasets(df4))
```
